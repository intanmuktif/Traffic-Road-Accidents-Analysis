## **4.1. Faktor utama yang mempengaruhi tingkat keparahan kecelakaan lalu lintas.**
1. Faktor utama yang mempengaruhi tingkat keparahan kecelakaan lalu lintas.
2. Apakah ada perbedaan signifikan tingkat keparahan kecelakaan berdasarkan faktor "Driving_experience", "Road_surface_type", "Light_conditions", dan "Weather_conditions"
(Statistika Inferensial)

```python
# Filter baris dengan accident severity bernilai 2
filtered_df = df[df['Accident_severity'] == 2]

# Memilih beberapa kolom yang memengaruhi keparahan kecelakaan
result_df = filtered_df[['Driving_experience', 'Road_surface_type', 'Light_conditions', 'Weather_conditions']]

# Reset index
result_df = result_df.reset_index(drop=True)

# Menampilkan statistika deskriptif untuk beberapa kolom tertentu
result_df.describe()
```

Menggunakan statistika deskriptif pada beberapa kolom yang menjadi faktor-faktor kecelakaan lalu lintas di jalan raya. Faktor-faktornya seperti pengalaman mengemudi, jenis permukaan jalan, kondisi cahaya, dan kondisi cuaca.<br>
Berdasarkan hasil statistika deskriptif pada kolom kategori 'Driving_experience', 'Road_surface_type', 'Light_conditions', dan 'Weather_conditions' dari 'Accident_severity' bernilai 2. Faktor utama yang mempengaruhi tingkat keparahan kecelakaan lalu lintas dari banyaknya frekuensi adalah dari jenis permukaan jalan yaitu jalan yang beraspal sebesar 6027 kasus. Namun kita perlu menganalisisnya lebih lanjut.

Selanjutnya melakukan **Uji Hipotesis :**<br>
Apakah ada perbedaan signifikan tingkat keparahan kecelakaan berdasarkan faktor "Driving_experience", "Road_surface_type", "Light_conditions", dan "Weather_conditions terhadap "Accident_Severity" ? <br>
* **Hipotesis Nol (H0)** = Tidak ada perbedaan signifikan dalam tingkat keparahan kecelakaan berdasarkan faktor tertentu.
* **Hipotesis Alternatif (H1)** = Ada perbedaan signifikan dalam tingkat keparahan kecelakaan berdasarkan faktor tertentu.

```python
# Uji ANOVA untuk setiap faktor
anova_results = {}
for col in ['Driving_experience', 'Road_surface_type', 'Light_conditions', 'Weather_conditions']:
    groups = [df[df[col] == level]['Accident_severity'] for level in df[col].unique()]
    anova_results[col] = f_oneway(*groups)

# Menampilkan hasil ANOVA
for factor, result in anova_results.items():
    print(f"Hasil ANOVA dari faktor {factor} terhadap Accident_Severity adalah: \nF-statistic = {result.statistic}, p-value = {result.pvalue}")
```

Dalam menjawab uji hipotesis apakah ada perbedaan signifikan tingkat keparahan kecelakaan berdasarkan faktor tertentu dapat menggunakan Uji Hipotesis Anova. Dalam hal ini, Anova membandingkan variabel antara kelompok (kategori dari faktor tertentu). Anova memungkinkan untuk membandingkan lebih dari dua kategori atau level dalam faktor yang diuji.<br>
F-statistic dalam Anova memberikan ukuran seberapa besar perbedaan antara grup dibandingkan dengan variabel acak di dalam grup. Ini adalah langkah pertama dalam menentukan apakah perbedaan yang diamati dalam data adalah signifikan atau hanya akibat variabel acak.<br>
Dalam uji Anova, nilai p-value adalah ukuran probabilitas yang digunakan untuk menentukan apakah ada perbedaan yang signifikan secara statistik antara rata-rata dari dua atau lebih grup yang diuji. P-value membantu kita menguji hipotesis nol (H0) terhadap hipotesis alternatif (H1).

```python
# Uji Hipotesis
alpha = 0.05
print("Uji Hipotesis:")
for col, result in anova_results.items():
    if result.pvalue < alpha:
        print(f'{col}: Ada perbedaan signifikan terhadap Accident Severity (Tolak H0)')
    else:
        print(f'{col}: Tidak ada perbedaan signifikan terhadap Accident Severity (Terima H0)')
```

Alpha adalah tingkat signifikansi yang digunakan dalam uji hipotesis. Biasanya, alpha ditetapkan pada 0.05 (5%), yang berarti kita menerima risiko 5% untuk melakukan kesalahan. Dalam hal ini, p-value adalah probabilitas yang dihitung dalam uji Anova. Ini menunjukkan seberapa konsisten data dengan hipotesis nol. Jika p-value kurang dari alpha (0.05), maka ada bukti yang cukup untuk menolak hipotesis nol (H0), yang berarti ada perbedaan signifikan antara rata-rata tingkat keparahan kecelakaan di berbagai level faktor tersebut.<br>
Jika p-value lebih besar atau sama dengan alpha, tidak ada bukti yang cukup untuk menolak hipotesis nol, yang berarti tidak ada perbedaan signifikan dalam tingkat keparahan kecelakaan berdasarkan faktor tersebut.

Berdasarkan hasil p-value uji Anova pada masing-masing kolom faktor yang memengaruhi keparahan kecelakaan, yang memiliki nilai p-value lebih dari 0.05 adalah dari kolom **Road_surface_type** sebesar 0.7362761567899162. Oleh karena itu, hasil uji hipotesisnya tidak ada perbedaan signifikan dalam tingkat keparahan kecelakaan berdasarkan faktor Road_surface_type / jenis permukaan jalannya, dalam hal ini H0 diterima.

Sementara, hasil uji hipotesis dari kolom Driving_experience, Light_conditions, dan Weather_conditions terhadap Accident_Severity memiliki p-value > 0.05. Artinya, ada perbedaan signifikan dalam tingkat keparahan kecelakaan berdasarkan faktor tersebut.

## **4.3. Waktu kecelakaan lalu lintas yang paling sering terjadi**

```python
# Tambahkan kolom baru untuk bulan
df['Bulan'] = df['Date'].dt.to_period('M')

# Hitung frekuensi kecelakaan per bulan
kecelakaan_per_bulan = df['Bulan'].value_counts().sort_index()

# Konversi index ke datetime untuk plotting
kecelakaan_per_bulan.index = kecelakaan_per_bulan.index.to_timestamp()

# Buat line chart menggunakan Plotly
fig = px.line(x=kecelakaan_per_bulan.index, y=kecelakaan_per_bulan.values,
              labels={'x': 'Bulan', 'y': 'Jumlah Kecelakaan'},
              title='Frekuensi Kecelakaan Lalu Lintas per Bulan')

# Sesuaikan format tanggal pada sumbu x
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y"
)

# Tampilkan grafik
fig.show()
```

 Dari line chart mengenai frekuensi kecelakaan lalu lintas per bulan, dapat diperoleh informasi sebagai berikut :
 1. Jumlah kecelakaan lalu lintas menunjukkan variabilitas yang signifikan setiap bulan. Ada bulan-bulan dengan jumlah kecelakaan yang tinggi, diikuti oleh penurunan tajam, dan kemudian peningkatan lagi.
 2. Tidak ada pola yang jelas. Kecelakaan tampaknya terjadi secara acak sepanjang tahun tanpa konsistensi yang dapat diidentifikasi seperti lebih tinggi di musim tertentu.
 3. Ada beberapa puncak yang menonjol, misalnya pada bulan Maret 2020, di mana jumlah kecelakaan mencapai angka tertinggi sebesar 185 kasus.
 4. Penurunan terbesar dalam jumlah kecelakaan terjadi pada bulan-bulan tertentu, misalnya, seperti di bulan Maret 2020 ke April 2020.
 6. Jumlah kecelakaan terendah terdapat 135 kasus pada bulan Februari 2017 dan Februari 2018

```python
# Tambahkan kolom baru untuk bulan
df['Bulan'] = df['Date'].dt.to_period('M')

# Hitung frekuensi kecelakaan per bulan untuk setiap tingkat keparahan
kecelakaan_per_bulan = df.groupby(['Bulan', 'Accident_severity']).size().reset_index(name='Jumlah Kecelakaan')

# Konversi index ke datetime untuk plotting
kecelakaan_per_bulan['Bulan'] = kecelakaan_per_bulan['Bulan'].dt.to_timestamp()

# Warna spesifik untuk setiap tingkat keparahan
color_discrete_map = {
    0: 'blue',   # Warna untuk tingkat keparahan 0
    1: 'green',  # Warna untuk tingkat keparahan 1
    2: 'red'     # Warna untuk tingkat keparahan 2
}

# Buat line chart menggunakan Plotly dengan warna berdasarkan Accident_severity
fig = px.line(kecelakaan_per_bulan, x='Bulan', y='Jumlah Kecelakaan', color='Accident_severity',
              labels={'Bulan': 'Bulan', 'Jumlah Kecelakaan': 'Jumlah Kecelakaan', 'Accident_severity': 'Tingkat Keparahan'},
              title='Frekuensi Kecelakaan Lalu Lintas per Bulan berdasarkan Tingkat Keparahan',
              color_discrete_map=color_discrete_map)

# Sesuaikan format tanggal pada sumbu x
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y"
)

# Tampilkan grafik
fig.show()
```

Informasi yang dapat diambil dari frekuensi kecelakaan lalu lintas per bulan berdasarkan tingkat keparahan menggunakan multiple line chart, antara lain :
1. Grafik ini menunjukkan berapa banyak kecelakaan lalu lintas yang terjadi setiap bulan dari tahun 2017 sampai 2020.

## **4.6. Hubungan faktor usia mempengaruhi tingkat keparahan kecelakaan**
Bagaimana faktor kelompok usia pengemudi mempengaruhi tingkat keparahan kecelakaan? (Statistika Deskriptif)

```python
# Uji Chi-Square
chi2_results = {}
for col in ['Age_band_of_driver']:
    contingency_table = pd.crosstab(df[col], df['Accident_severity'])
    chi2_results[col] = stats.chi2_contingency(contingency_table)

# Menampilkan hasil Chi-Square
for factor, result in chi2_results.items():
    chi2, p, dof, ex = result
    print(f"Hasil uji korelasi Chi-Square dari faktor {factor} terhadap Accident_severity adalah:\nChi2 = {chi2}, p-value = {p}")
```

Pada permaalahan ini, Uji Chi-square digunakan untuk menentukan apakah ada hubungan antara dua variabel kategori.<br>
Hasil ini menunjukkan bahwa terdapat hubungan yang signifikan secara statistik antara kelompok usia pengemudi dan tingkat keparahan kecelakaan. P-value sebesar 0.0000823, yang jauh lebih kecil dari 0.05, menunjukkan bahwa kelompok usia pengemudi berkaitan dengan tingkat keparahan kecelakaan.<br>
Tingkat keparahan kecelakaan dapat dikaitkan dengan perbedaan dalam kelompok usia pengemudi, menunjukkan bahwa faktor usia pengemudi berperan penting dalam menentukan seberapa parah kecelakaan itu.
